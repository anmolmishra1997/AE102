```python
f = open("C:\Users\Anmol Mishra\Desktop\AE102\sslc1.txt")
data = f.read()
f.close()
data = data.split('\n')
del data[-1]
```

```python
region = {}
for line in data:
    fields = line.split(';')
    char = fields[0]
    if char in region:
        region[char] += 1
    else:
        region[char]= 1
print(region)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.pie(region.values(), explode = [0.05,0.05,0.05,0.05,0.05,0.05], labels = region.keys(), autopct = '%.2f')
print ( "Data values shown are the percentage of students from each region")
```

```python
pas = dict.fromkeys(region,0)
for line in data:
    fields = line.split(';')
    char = fields[0]
    passstatus = fields[9]
    if (passstatus == 'P'):
        pas[char]+=1
```

```python
import numpy as np
passNumber = np.array(pas.values())
regionNumber = np.array(region.values())
passPercentage = 100.0 * passNumber/regionNumber
print (passPercentage)
```

```python
plt.pie(passPercentage, labels =pas.keys(),autopct = lambda(p): '{:.02f}'.format(p * sum(passPercentage) / 100),explode = [0.05,0.05,0.05,0.05,0.05,0.05])
print ("Data values shown are the actual percentage of students passing in each region")
```

```python
presentAll = dict.fromkeys(region,0)
for line in data:
    fields = line.split(';')
    char = fields[0]
    if ('AA' in fields):
        presentAll[char]+=1
presentAllNumber = np.array(presentAll.values())
presentAllNumber = regionNumber - presentAllNumber
presentAllPercentage = 100.0 * presentAllNumber/regionNumber
print(presentAllPercentage)
```

```python
plt.pie(presentAllPercentage,labels = presentAll.keys(),1, explode = [0.05,0.05,0.05,0.05,0.05,0.05])
print ("Data values shown are the actual percentage of students present in all exams in each region")
```

```python
sciencemarks = dict.fromkeys(region,0)
for line in data:
    fields = line.split(';')
    char = fields[0]
    marks = fields[6]
    if 'AA' in marks:
        marks=0
    else:
        marks=int(marks)
    if marks>90:
        sciencemarks[char] += 1
```

```python
sciencemarksNumber = np.array(sciencemarks.values())
sciencePercentage = 100.0 * sciencemarksNumber/regionNumber
print(sciencePercentage)
```

```python
plt.pie(sciencePercentage,labels = sciencemarks.keys(),autopct = lambda(p): '{:.02f}'.format(p*sum(sciencePercentage)/100),explode = [0.05,0.05,0.05,0.05,0.05,0.05])
print ("Data values shown are the actual percentage of students scoring above 90 percent in Science in each region")
```
