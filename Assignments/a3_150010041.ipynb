```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\Users\Anmol Mishra\Desktop\AE102\Assignments\sslc1.csv", sep = ';')
dffinal = pd.read_csv("C:\Users\Anmol Mishra\Desktop\AE102\Assignments\sslc1.csv", sep = ';', na_values=['AA'])
```

```python
dfA = pd.DataFrame()
dfB = pd.DataFrame()
dfC = pd.DataFrame()
dfD = pd.DataFrame()
dfE = pd.DataFrame()
dfF = pd.DataFrame()
regionNumber = df.region.value_counts()
```

```python
df.region.value_counts().plot.pie(explode = [0.05,0.05,0.05,0.05,0.05,0.05], autopct = '%.2f')
plt.title ("Pie chart depicting percentage of students in each region")
```

```python
df_pass = df[df['pass'] == 'P']
passNumber = df_pass.region.value_counts()
passPercentage = 100.0 * passNumber/regionNumber
passPercentage.plot.pie(autopct = lambda(p): '{:.02f}'.format(p * sum(passPercentage) / 100), explode = [0.05,0.05,0.05,0.05,0.05,0.05])
plt.title("Percentage of students passing from each region")
```

```python
df_allPresent = df[(df.sl != 'AA') & (df.fl!= 'AA') & (df.math!= 'AA') & (df.sci!='AA') & (df.ss!='AA')]
presentAllNumber = df_allPresent.region.value_counts()
presentAllPercentage = 100.0 * presentAllNumber/regionNumber
presentAllPercentage.plot.pie(autopct = lambda(p): '{:.02f}'.format(p*sum(presentAllPercentage)/100), explode = [0.05,0.05,0.05,0.05,0.05,0.05])
plt.title("Percentage of students present in all exams from each region")
```

```python
dffinal.plot.hist(y='total',bins = 50)
plt.title ("Histogram of total marks for students of all regions")
```

```python
dfarr = [dfA,dfB,dfC,dfD,dfE,dfF]
region = ['A','B','C','D','E','F']
for i in range(6):
    dfarr[i] = dffinal[(dffinal.region == region[i])]
    dfarr[i].plot.hist(y='total',bins =50)
    plt.title("Histogram for total marks of all students in region " + region[i])
```

```python
dffinal.plot.box(y='math')
plt.title ("Box plot of math marks for students of all regions")
```

```python
dfarr = [dfA,dfB,dfC,dfD,dfE,dfF]
region = ['A','B','C','D','E','F']
for i in range(6):
    dfarr[i] = dffinal[dffinal.region == region[i]]
    dfarr[i].plot.box(y='math')
    plt.title("Box plots for math marks of all students in region " + region[i])
```

```python
sciencemarksNumber = dffinal[dffinal.sci>90].region.value_counts()
sciencePercentage = 100.0 * sciencemarksNumber/regionNumber
sciencePercentage.plot.pie(autopct = lambda(p): '{:.02f}'.format(p*sum(sciencePercentage)/100),explode = [0.05,0.05,0.05,0.05,0.05,0.05])
plt.title("Percentage of students scoring above 90 in science in each region")
```
